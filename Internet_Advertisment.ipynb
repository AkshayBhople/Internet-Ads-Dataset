```python
# Converting 'ad.' to 1 & 'nonod.' to 0 in the last column
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lb = LabelEncoder()
y = lb.fit_transform(data.iloc[:, -1])
y
```

```python
from sklearn.utils import shuffle
data=shuffle(data)
data
```

```python
target=y
inputs=data.drop(['1558'],axis=1)
```

```python
# Scaling the inputs
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
X = scaler.transform(inputs)
```

```python
# Splitting the data set into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=7)
```

```python
#creating function for model selection by using cross validation
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score

def test_model(model, X_train=x_train, y_train=y_train):
    cv = KFold(n_splits = 3, shuffle=True, random_state = 45)
    #r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, X_train, y_train, cv=cv)
    score = [r2_val_score.mean()]
    return score
```

```python
from sklearn.svm import SVC
clf=SVC(kernel='rbf',random_state=5,C=100)
test_model(clf)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
test_model(clf1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf2 = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p = 2)
test_model(clf2)
```

```python
from sklearn.linear_model import LogisticRegression
clf3 = LogisticRegression(random_state = 0)
test_model(clf3)
```

```python
from sklearn.naive_bayes import GaussianNB
clf4 = GaussianNB()
test_model(clf4)
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier,BaggingClassifier,ExtraTreesClassifier
clf5 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
test_model(clf5)
```

```python
clf6=AdaBoostClassifier()
test_model(clf6)
```

```python
clf7=GradientBoostingClassifier()
test_model(clf7)
```

```python
clf8=BaggingClassifier()
test_model(clf8)
```

```python
clf9=ExtraTreesClassifier()
test_model(clf9)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
import time
start_time = time.time()
clf7.fit(x_test,y_test)
print("model fit time --> {}".format(time.time()-start_time))
start_time = time.time()
y_pred=clf7.predict(x_test)
print("model predict time --> {}".format(time.time()-start_time))
print('The accuracy of the model is :',accuracy_score(y_test,y_pred))
print('The f1_score of the model is :',f1_score(y_test, y_pred, average = 'binary'))
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
params = {'n_estimators':[50,70,100,120]
         }
rand_search = RandomizedSearchCV(clf7, param_distributions=params, n_jobs=-1, cv=11)
rand_search.fit(x_train, y_train)
rand_search.best_score_
```
